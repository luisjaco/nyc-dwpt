## Overview
Here we analyze the outcomes of our 42nd St. simulation. This simulation left traffic under
consistent scenarios as the previous, however only vehicles which traveled the entirety of 42nd St.
are included here.

## Setup

```python
import pandas as pd
import numpy as np
import scipy.stats as scipy
import matplotlib.pyplot as pyplot
import seaborn
```

```python
def perform(df_list, func): 
    for df in df_list:
        func(df)
```

```python
df = pd.read_csv('data/nyc42nd.csv')
display(df)
```

## Feature Engineering

```python
# create in hours column
df['total_minutes'] = df['total_seconds'] / 60
df['total_hours'] = df['total_seconds'] / 3600
display(df)
```

## Statistical Analysis

```python
# lets check statistics
df.describe()
```

```python
mean = np.mean(df.total_minutes)
pyplot.hist(data=df, x='total_minutes', bins=20)
pyplot.title(f'Total time spent crossing 42nd Street per vehicle\nAverage: {mean:.2f} minutes')
pyplot.axvline(x=mean).set_color('pink')
pyplot.xlabel('Time (minutes)')
pyplot.ylabel('Number of Vehicles')
pyplot.plot()
```

## Conclusions
Visually it appears that, with the simulated data, there is a bell curve, or normal distribution, 
about the mean. The simulations suggest a trip time of around 40.64 minutes to cross the entirety of 
42nd St. within a medium-traffic scenario.
